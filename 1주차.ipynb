```python
movie_list=["주토피아","아바타","도라에몽","오버워치","동물의숲","going home","내가 제일 잘 나가","개발자란","d-2","내 휴일이 어디갔지"]
age_list=[]
adult=0
child=0

for i in range(10):
  print(f"영화 {i} : {movie_list[i]}")
movie_choice=int(input("영화를 선택하세요 : "))
print(f"{movie_list[movie_choice]} 선택했습니다.")

print("----------------------------------")

num_people=int(input("인원 수 입력 : "))
print(f"{num_people}명 선택했습니다.")

print("----------------------------------")

input_age=input("나이를 입력하세요 : ").split()
for age in input_age:
  age_list.append(int(age))

for i in range(num_people):
  if age_list[i]<20:
    child += 1
  else:
    adult += 1
print(f"어른 : {adult}명, 아이 : {child}명 선택했습니다.")

print("----------------------------------")

seat=input("좌석 선택(A~Z)")
print(f"{seat}좌석 선택했습니다.")

print("----------------------------------")

price=adult*10000+child*5000
print(f"{movie_list[movie_choice]}, {num_people}명, {age_list}세, 좌석 : {seat} 금액 : {price}")
while(1):
  input_money=int(input("돈을 넣어주세요 : "))
  if(input_money<price):
    print("돈이 부족합니다.")
  else:
    change=input_money-price
    print(f"거스름돈 : {change}")
    break;
print("결제가 완료되었습니다.")

print("----------------------------------")
print("즐거운 관람되세요.")


```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee(Person):
    def payment(self):
        print("결제가 완료되었습니다.")
        print("----------------------------------")
        print("즐거운 관람되세요.")

class Movie:
    def __init__(self, movie_list):
        self.movie_list = movie_list

    def show_list(self):
        for i in range(len(self.movie_list)):
            print(f"영화 {i} : {self.movie_list[i]}")

class Ticket:
    def __init__(self, movie_name, num_people, age_list, seat):
        self.movie_name = movie_name
        self.num_people = num_people
        self.age_list = age_list
        self.seat = seat
        self.adult = 0
        self.child = 0
        self.price = self.calculate()

    def calculate(self):
        for age in self.age_list:
            if age < 20:
                self.child += 1
            else:
                self.adult += 1
        return self.adult * 10000 + self.child * 5000

movie_list = ["주토피아","아바타","도라에몽","오버워치","동물의숲","going home","내가 제일 잘 나가","개발자란","d-2","내 휴일이 어디갔지"]
movie_mgr = Movie(movie_list)
movie_mgr.show_list()

movie_choice = int(input("영화를 선택하세요 : "))
print(f"{movie_list[movie_choice]} 선택했습니다.")

print("----------------------------------")

num_people = int(input("인원 수 입력 : "))
print(f"{num_people}명 선택했습니다.")

print("----------------------------------")

input_age = input("나이를 입력하세요 : ").split()
age_list = []
for age in input_age:
    age_list.append(int(age))

seat = input("좌석 선택(A~Z)")
print(f"{seat}좌석 선택했습니다.")

print("----------------------------------")

ticket = Ticket(movie_list[movie_choice], num_people, age_list, seat)
print(f"어른 : {ticket.adult}명, 아이 : {ticket.child}명 선택했습니다.")

print("----------------------------------")

print(f"{ticket.movie_name}, {ticket.num_people}명, {ticket.age_list}세, 좌석 : {ticket.seat} 금액 : {ticket.price}")

staff = Employee("직원", 20)

while True:
    input_money = int(input("돈을 넣어주세요 : "))
    if input_money < ticket.price:
        print("돈이 부족합니다.")
    else:
        change = input_money - ticket.price
        print(f"거스름돈 : {change}")
        staff.payment()
        break
```
